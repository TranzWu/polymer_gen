```python
import jax
import jax.numpy as jnp
from itertools import combinations
import numpy as np
```

```python
def potential(x):
    pairs = [comb for comb in combinations(x, 2)]
    pot = sum([(p[0]-p[1])**(-12) - (p[0]-p[1])**(-6) for p in pairs])
    return -pot

potential_grad = jax.grad(potential)
```

```python
x = np.array([0., 0.85, 1.68, 3., 4.])
x = jnp.array(x)
```

```python
x
```

```python
traj = []
traj.append(x)
for i in range(10):
    grad_x = potential_grad(x)
    x += 0.001 * grad_x
    traj.append(x)
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(len(traj)):
    plt.scatter(traj[i], [i for j in range(len(traj[i]))], label=f'n_iter={i}')
plt.legend()
plt.savefig('1d')
```
